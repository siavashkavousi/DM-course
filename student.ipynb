```python
# Remove samples with 5 cols of zero
train_df = train_df.loc[(train_df.iloc[:, :4] != 0).any(axis=1)]
```

```python
train_df[train_df == 0] = np.nan
train_df.head()
```

```python
g0_count = train_df['g0'].isnull().sum()
g1_count = train_df['g1'].isnull().sum()
g2_count = train_df['g2'].isnull().sum()
g3_count = train_df['g3'].isnull().sum()
g4_count = train_df['g4'].isnull().sum()
g5_count = train_df['g5'].isnull().sum()

g0_count, g1_count, g2_count, g3_count, g4_count, g5_count
```

```python
from sklearn.model_selection import train_test_split 

n_samples = train_df.shape[0]
n_features = train_df.shape[1]
x = train_df.iloc[:, :6]
y = train_df.iloc[:, 6]

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2)
```

```python
from sklearn.preprocessing import Imputer
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor
from sklearn.pipeline import Pipeline
from sklearn.model_selection import cross_val_score

estimator = Pipeline([("imputer", Imputer(missing_values=0,
                                          strategy="mean",
                                          axis=0)),
                      ("forest", RandomForestRegressor(random_state=0,
                                                       n_estimators=100))])
```

```python
from sklearn.model_selection import cross_val_score, KFold
from sklearn.metrics import mean_squared_error, make_scorer

kf = KFold(n_splits=10)
scores = cross_val_score(estimator,
                         x_train,
                         y_train,
                         cv=kf,
                         scoring=make_scorer(mean_squared_error))
np.sqrt(scores).mean()
```
